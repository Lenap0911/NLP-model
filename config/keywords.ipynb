A notebook to further identify and refine search keywords for the flood NLP pipeline.

```python
import re 

import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
#load txt file
with open('en_search_keywords.txt', 'r') as file:
    full_text = file.read().splitlines()

```

```python
#clean text
#1- seperate into sentences
sentences = []
for text in full_text:
    sentences.extend(re.split(r'(?<=[.!?]) +', text))

#2- remove empty sentences
sentences = [s for s in sentences if s.strip()]


#print sentence one by one
for sentence in sentences:
    print(sentence)
```

```python
# identify the verb in each sentence and count how many times each verb appears
verbs_count = {}
for sentence in sentences:
    doc = nlp(sentence)
    # make verbs into simplest form
    verbs = [token.lemma_ for token in doc if token.pos_ == "VERB"]
    #append verb if not already there
    for v in verbs:
        if v not in verbs_count:
            verbs_count[v] = 0
        verbs_count[v] += 1

# print verbs from highest to lowest count
sorted_verbs = sorted(verbs_count.items(), key=lambda x: x[1], reverse=True)
for verb, count in sorted_verbs:
    print(f"{verb}: {count}")
```
